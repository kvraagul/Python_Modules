```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

```python
df=sns.load_dataset('tips')
```

```python
df.head() 
```

# Line Plot

```python
X = np.linspace(1,20,100)
Y = np.exp(X)

plt.plot(X,Y)
plt.show()
```

```python
X = np.linspace(1,20,100)
Y = np.exp(X)

sns.lineplot(X,Y)
```

```python
X = np.linspace(1,20,100)
Y = np.exp(X)

plt.plot(X,Y,'g*')
plt.show()
```

```python
X = np.linspace(1,20,100)
Y = np.exp(X)
Z = np.sqrt(X)

plt.plot(X,Y,'r',X,Z,'b')
plt.show()
```

# Scatter Plot

```python
A = df['total_bill']
B = df['tip']

plt.scatter(A,B)
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.scatterplot(df['total_bill'],df['tip'])
```

# Bar Plot

Bar Plot is used for categorical data

```python
smoker_value = df['smoker'].value_counts()
smoker_value.plot(kind='bar',rot=0) # rot - rotation

plt.xlabel('smoker')
plt.ylabel('smoker count')
```

# Pie Plot

```python
df['day'].value_counts()
```

```python
plt.pie(df['day'].value_counts(),labels=['sat','sun','thur','fri'],autopct='%.0f%%')
plt.show()
```

```python
plt.pie(df['day'].value_counts(),labels=['sat','sun','thur','fri'],autopct='%.0f%%',explode=[0,0,0,0.5])
plt.show()
```

# Histogram Plot

A histogram is the bar chart for numerical data

```python
plt.hist(df['tip'])
```

```python
plt.hist(df['tip'],bins=5)
plt.show()
```

```python
sns.histplot(df['tip'])
```

```python
sns.histplot(df['tip'],bins=5,kde=True)
```

# Box Plot

used to find outliers

```python
plt.boxplot(df['total_bill'])
plt.show()
```

```python
plt.boxplot(df['total_bill'])

plt.text(x=1.1,y=df['total_bill'].min(),s='min')
plt.text(x=1.1,y=df['total_bill'].max(),s='max')
plt.text(x=1.1,y=df['total_bill'].median(),s='median')

plt.text(x=1.1,y=df.total_bill.quantile(0.25),s='0.25(Q1)')
plt.text(x=1.1,y=df.total_bill.quantile(0.75),s='0.75(Q1)')

plt.show()
```

```python
sns.boxplot(df['total_bill'])
```

```python
sns.boxplot(df['day'],df['total_bill'])
```

```python
sns.boxplot(data=df,x='day',y='total_bill',hue='smoker')
```

# Strip Plot

```python
sns.stripplot(data=df,x='time',y='tip')
```

# Swarm Plot

```python
sns.swarmplot(data=df,x='time',y='tip')
```

```python
sns.swarmplot(data=df,x='time',y='tip',hue='day')
```

# Violin Plot

```python
sns.violinplot(data=df,x='time',y='tip')
```

```python
sns.violinplot(data=df,x='day',y='total_bill',hue='sex',split=True)
```

# Pair Plot

```python
sns.pairplot(df)
```

# Distribution Plot

```python
sns.distplot(df['total_bill'])
```

```python
time=['Lunch','Dinner']

for i in time:

  subset=df[df['time']==i]

  sns.distplot(subset['total_bill'],hist=False,kde=True,kde_kws={'shade':True},label=i)
```

# Count Plot

count plot is similar to bar plot with multiple dimensions

```python
sns.countplot(data=df,x='day',hue='time')
```

# Heat Map

it is used to find correlation

```python
correlation = df.corr()
correlation
```

```python
sns.heatmap(correlation,annot=True)
```
