## datetime.date()

with datetime.date() we can work with years, months and day

**datetime.date(year,month,day)**

**Shouldn't add 0 in dates or months**

```python
d = datetime.date(2020,02,07)
print(d)
```

```python
d = datetime.date(2020,2,7)
print(d)
```

```python
d = datetime.date.today()
print(d)
```

**weekday**

  monday - 0
  sunday - 6

**isoweekday**

monday - 1
sunday - 7

```python
d = datetime.date.today()
print(d.day)
print(d.weekday())
print(d.isoweekday())
```

**timedelta**

used for calculating differences in dates and also can be used for date manipulations

```python
d = datetime.date.today()
print(d)
td = datetime.timedelta(days=10)
print(td)
print(td + d)
```

```python
today = datetime.date.today()
bday = datetime.date(2022,2,7)
till_bday = bday - today
print(till_bday)
print(till_bday.days)
```

## datetime.time()

with datetime.time() we can work with hours, minutes and seconds

**datetime.time(hours,minutes,seconds,microseconds)**

```python
t = datetime.time(16,45,30,10000)
print(t)
print(t.hour)
```

## datetime.datetime()

with datetime.datetime() we can work with year,months,day,hours, minutes and seconds

**datetime.time(year,month,day,hours,minutes,seconds,microseconds)**

```python
dt = datetime.datetime.today()
print(dt)
print(dt.date())
print(dt.time())
```

```python
dt = datetime.datetime(2050,3,5,18,39,29,100)
print(dt)
```

```python
dt = datetime.datetime.today()
print(dt)
td = datetime.timedelta(days = 365)
oneyear_from_now = td + dt
print(oneyear_from_now)
```

```python
dt_today = datetime.datetime.today()
dt_now = datetime.datetime.now()
dt_utcnow = datetime.datetime.utcnow()
print(f'dt_today : {dt_today}')
print(f'dt_now : {dt_now}')
print(f'dt_utcnow : {dt_utcnow}')
```

# pytz

**For timezones it is recommended to work with pytz**

```python
import pytz
```

```python
for i in dir(pytz):
  print(i)
```

```python
dt = datetime.datetime(2050,3,5,18,39,29,100000,tzinfo=pytz.UTC)
print(dt)
```

```python
dt_now = datetime.datetime.now()
print(f'dt_now : {dt_now}')
dt_now = datetime.datetime.now(tz=pytz.UTC)
print(f'dt_now pytz: {dt_now}')
```

```python
dt_now = datetime.datetime.now(tz=pytz.UTC)
my_time = dt_now.astimezone(pytz.timezone('Asia/Kolkata'))
print('kolkata :',my_time)
my_time2 = dt_now.astimezone(pytz.timezone('US/Mountain'))
print('US Mountain :',my_time2)
```

```python
for i in pytz.all_timezones:
  print(i)
```
