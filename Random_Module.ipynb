# Random Module

Python has a built-in module that you can use to make random numbers.

```python
import random
```

```python
value = random.random()
print(value)
```

```python
value = random.uniform(1,10)
print(value)
```

```python
value = random.randint(1,10)
print(value)
```

```python
greetings = ['hey','hello','hola','bonjur','namaste','vanakam']

value = random.choice(greetings)
print(value, 'Raagul')
```

```python
colors = ['red','black','blue','white','green']

result = random.choices(colors,k=8)
print(result)
```

```python
num =list(range(1,20))

random.shuffle(num)
print(num)
print('----------------------------------------------------------------------------------')
val = random.sample(num,k=7)
print(val)
```
